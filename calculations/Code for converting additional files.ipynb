Выгрузка с API data.mos информации об Объектах ритуального обслуживания.

```python
key = '?api_key=<ваш ключ>'
url = 'https://apidata.mos.ru/v1/datasets/607/features' 

ritual_df = []

try:
    response = requests.get(url+key)
    json_response = response.json()

    ritual_df += json_response['features']
except Exception as ex:
    print(ex)
    
```

```python
ritual_dataset = pd.DataFrame(ritual_df)
ritual_dataset_open = ritual_dataset["properties"].apply(pd.Series)
ritual_name = ritual_dataset_open["Attributes"].apply(pd.Series) 
display(ritual_name)
```

Редактирование колонок ['District'] и ['AdmArea'], заполнение пропусков.

```python
ritual_name['AdmArea'] = ritual_name['AdmArea'].str.replace('ё', 'е')
ritual_name['AdmArea'] = ritual_name['AdmArea'].fillna('нет информации')

ritual_name['District'] = ritual_name['District'].str.replace('район ', '').str.replace('поселение ', '').str.replace('ё', 'е').str.replace(' район', '').str.replace('"', '')
ritual_name['District'] = ritual_name['District'].fillna('нет информации')

display(ritual_name)
```

Соединение колонок ['Longitude_WGS84'] и ['Latitude_WGS84'] для дальнейшего удобства.

```python
ritual_name.Longitude_WGS84 = ritual_name.Longitude_WGS84.astype(str).str.cat(ritual_name.Latitude_WGS84, sep=',')
ritual_name = ritual_name.drop('Latitude_WGS84', axis=1)
ritual_name = ritual_name.rename({'Longitude_WGS84' : 'coordinates'}, axis=1)
display(ritual_name)
```
