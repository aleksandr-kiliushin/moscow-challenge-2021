Группировка данных по году и типу учреждения, изменение типов колонок.

```python
plan_schools_сsv = pd.read_csv('plan_schools.csv')
plan_schools_grouped = plan_schools_сsv.groupby(['ReportingYear', 'LaunchedEducationalFacilityType']).agg(sum).drop(['Unnamed: 0', 'global_id'], axis=1)
plan_schools_grouped['LaunchedFacilitiesPlannedQuantityYear'] = plan_schools_grouped['LaunchedFacilitiesPlannedQuantityYear'].astype(int)
plan_schools_grouped['LaunchedFacilitiesActualQuantityReportPeriod'] = plan_schools_grouped['LaunchedFacilitiesActualQuantityReportPeriod'].astype(int)
plan_schools_grouped['ActualVacanciesQuantityReportPeriod'] = plan_schools_grouped['ActualVacanciesQuantityReportPeriod'].astype(int)
plan_schools_grouped['PlannedVacanciesQuantityYear'] = plan_schools_grouped['PlannedVacanciesQuantityYear'].astype(int)
display(plan_schools_grouped)
```

Выявление недостроя количества мест.

```python
plan_schools_grouped['ShortageFacilitiesQuanityYear'] = plan_schools_grouped['LaunchedFacilitiesPlannedQuantityYear'] - plan_schools_grouped['LaunchedFacilitiesActualQuantityReportPeriod']
plan_schools_grouped['ShortageVacanciesQuanityYear'] = plan_schools_grouped['PlannedVacanciesQuantityYear'] - plan_schools_grouped['ActualVacanciesQuantityReportPeriod']
plan_schools_grouped = plan_schools_grouped.reset_index()
display(plan_schools_grouped)
```

Расчет индекса недостроя мест.

```python
def shortage_percent(planned, shortage):
    percent = round((shortage * 100) / planned, 2)
    print(percent)

shortage_vacancies_percent_year_preschool = shortage_percent(plan_schools_grouped[plan_schools_grouped['LaunchedEducationalFacilityType'] == 'Дошкольные образовательные учреждения']['PlannedVacanciesQuantityYear'].sum(), plan_schools_grouped[plan_schools_grouped['LaunchedEducationalFacilityType'] == 'Дошкольные образовательные учреждения']['ShortageFacilitiesQuanityYear'].sum())
shortage_vacancies_percent_year_school = shortage_percent(plan_schools_grouped[plan_schools_grouped['LaunchedEducationalFacilityType'] == 'Школы и блоки начальных классов']['PlannedVacanciesQuantityYear'].sum(), plan_schools_grouped[plan_schools_grouped['LaunchedEducationalFacilityType'] == 'Школы и блоки начальных классов']['ShortageFacilitiesQuanityYear'].sum())
```
